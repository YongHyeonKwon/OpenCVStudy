```python
import cv2
img_color = cv2.imread('./data/pika.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.imread('./data/pika.jpg', cv2.IMREAD_GRAYSCALE)
img_unch = cv2.imread('./data/pika.jpg', cv2.IMREAD_UNCHANGED)

cv2.imshow('img_color', img_color)
cv2.imshow('img_gray', img_gray)
cv2.imshow('img_unch', img_unch)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 이진(binary) 이미지 생성 
 - 픽셀값의 의미가 0과 1로 구성
 - cv2.threshold(img, threshold_value, value, flag)
  - img : Grayscale 이미지
  - threshold_value : 픽셀 문턱값 (0과 1로 나누기 위한 기준값)
  - value : 문턱값보다 클 때 적용되는 값
  - flag : 문턱값 적용 방법
    - cv2.THRESH_BINARY : 픽셀값이 문턱값보다 크면 value, 아니면 0을 할당
    - cv2.THRESH_BINARY_INV : 픽셀값이 문턱값보다 크면 0, 아니면 value를 할당  

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('./data/pika.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray')

# 이진화 진행하는 함수 
# 130보다 크면 흰색, 130보다 작으면 검은색
_, thr_img = cv2.threshold(img, 130,255, cv2.THRESH_BINARY)
plt.imshow(thr_img, cmap = 'gray')
thr_img
```

#### 동영상 출력 

```python
import cv2
cap = cv2.VideoCapture('./data/car_accident.mp4')
while cap.isOpened(): # 동영상 파일이 바르게 열렸는지?
    ret, frame = cap.read() # ret: 성공여부, frame : 받아온 이미지(프레임)
    if not ret:
        print('동영상 재생 끝')
        break
        
    cv2.imshow('video', frame)
    if cv2.waitKey(30) == ord('q'):
        print('사용자 입력에 의해서 종료')
        break
cap.release() # 카메라 , 동영상 지원 해제
cv2.destroyAllWindows() # 모든 윈도우 창 끄기
```

```python
import cv2

try:
    cap = cv2.VideoCapture(0)
    print('비디오 캡처 시작')    
except:
    print("비디오 캡처 실패")

# 파일명에 숫자을 붙이기 위한 변수 
co = 0

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("비디오 읽기 실패 또는 비디오 모두 읽음")
        cap.release()
        cv2.destroyAllWindows()
        break

    cv2.imshow("frame", frame)        
    k = cv2.waitKey(30)
    
    if k == ord('q'):
        co = co + 1
        print("./data/pic{}.png  파일을 저장".format(co))
        cv2.imwrite("./data/pic{}.png".format(co), frame, 
                     params=[cv2.IMWRITE_PNG_COMPRESSION, 0])

    if k == ord('w'):
        cap.release()
        cv2.destroyAllWindows()
        break
```

```python
import cv2
cap = cv2.VideoCapture(0) # 0번째 카메라 장치( Device ID , 실행창 devmgmt.msc>카메라>속성)

# 코덱 정의
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps =  30.0  # 초당 프레임 이미지 수

# 저장 파일명, 코덱, fps, 크기(가로, 세로)
out = cv2.VideoWriter('out.avi', fourcc, fps, (width, height))

# 녹화 상태 여부
record = False

if not cap.isOpened(): # 카메라가 잘 열리지 않은 경우
    exit() # 프로그램 종료
    
while True:
    ret, frame = cap.read()
    if not ret:
        print("비디오 읽기 실패 또는 비디오 모두 읽음")
        break
    out.write(frame)
    cv2.imshow('camera', frame)
    
    if record :
        out.write(frame)
        
    if cv2.waitKey(30) == ord('q'): # 사용자가 알파벳 q를 눌렀을 때 카메라 끄기
        print('카메라 끄기')
        out.release()
        cap.release()
        cv2.destroyAllWindows()
        break
    elif cv2.waitKey(30) == ord('w'):
        print('녹화 시작!')
        record = True
```
